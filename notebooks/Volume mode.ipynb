```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv("/home/thierry/Documents/thierry/data/user1.csv")
data['index'] = range(0, len(data))
for i in range(len(data)):
    data['date'][i] = datetime.datetime.strptime(data['date'][i], '%a %b %d %H:%M:%S GMT-03:00 %Y').day
    
data['index'] = range(0, len(data))
data = data.query('date < 3 or date > 28')
```

```python
data30s = data.iloc[ 0::2 , [14,16]]
data60s = data.iloc[ 0::4 , [14,16]]
data120s = data.iloc[ 0::8 , [14,16]]
```

```python
def find_audio_mode(df, mode):
    usage = 0
    for i in df.index.values:
        if(df['audio'][i]==mode):
            usage+=1
        
    return usage
```

```python
n_vibrate = find_audio_mode(data30s, 'Vibrate')
n_silent = find_audio_mode(data30s, 'Silent')
n_normal = find_audio_mode(data30s, 'Normal')
```

```python
def find_all_audio_mode(df,grouping,mode):
    usage = 0
    for i in range(0,len(df),grouping):
        usage += checkMode(df[i:(i+grouping)],mode)
    return usage
```

```python
def checkMode(df,mode):
    result = False
    for i in df.index.values:
        result = result or (df['audio'][i] == mode)
    return 1 if result else 0 
```

```python
print("Porcentagem mode normal 30s",find_audio_mode(data30s,'Normal')/find_all_audio_mode(data,2,'Normal'))
print("Porcentagem mode normal 60s",find_audio_mode(data60s,'Normal')/find_all_audio_mode(data,4,'Normal'))
print("Porcentagem mode normal 120s",find_audio_mode(data120s,'Normal')/find_all_audio_mode(data,8,'Normal'))
```
